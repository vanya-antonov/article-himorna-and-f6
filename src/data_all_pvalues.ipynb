```python
import os
import pandas as pd

from my_functions import makeExactFisherTestFromFiles
from my_constants import ALL_MARKS, DATA_DIR
```

```python
pvalue_df_list = []
for mark_name in [x[0] for x in ALL_MARKS]:
    mark_dir = os.path.join(DATA_DIR, "all_marks/", mark_name)
    confusion_matrix_fn = os.path.join(mark_dir, "our_fantom_fantom_aso_genes_confusion_matrix.tsv")
    genes_association_fn = os.path.join(mark_dir, "our_fantom_fantom_aso_genes_association.tsv")
    pvalue_df = makeExactFisherTestFromFiles(confusion_matrix_fn, genes_association_fn)
    pvalue_df.insert(2, column="mark_name", value=mark_name)
    pvalue_df_list.append(pvalue_df)

all_pvalue_df = pd.concat(pvalue_df_list)
all_pvalue_df.to_csv(os.path.join(DATA_DIR, "all_pvalues.tsv"), sep="\t", index=False)
all_pvalue_df
```
